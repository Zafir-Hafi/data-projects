```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/PMLS/Downloads/Global Smartphone Shipments By Company By Quarter (Samsung Apple Huawei...).csv')


data = data.dropna()
data = data.drop_duplicates()

print(data)



x = data['Quarter']  # This should be the column name for the smartphone companies
y = data['Samsung']  # This should be the column name for the number of units shipped

# Create the bar plot
plt.bar(x, y, edgecolor='black')
plt.xticks(rotation=45)
plt.tight_layout()
# Add labels and title if necessary
plt.xlabel('Qaurter')
plt.ylabel('Samsung')

plt.title('Global Smartphone Shipments by Company')
plt.tight_layout()

# Show the plot
plt.show()
```
